## Data Cleaning 

```python
missing = pd.DataFrame((data.isnull().sum())*100/data.shape[0],columns=['Percentage']).reset_index()
missing.head()
```

```python
plt.figure(figsize=(20,5))
ax = sns.pointplot(data=missing,x="index",y="Percentage",capsize=5)
plt.xticks(rotation =90,fontsize =20)
plt.title("Percentage of Missing values")
plt.ylabel("Percentage",labelpad=10)
plt.xlabel("Index",labelpad=10)
plt.show()
```

As we can see that the missing values are less than 5% in most of the cases greater than 5% in just one case.
None of those columns are categorical so we can fill them with the mean values.

We make a copy of data inorder to manipulate and process it.

```python
copy=data.copy()
```

```python
copy=pd.get_dummies(data)
copy.replace({True:1, False:0},inplace=True)
copy.head()
```

```python
copy.fillna(copy.mean(),inplace=True)
copy.head()
```

Divide customers into bins based on tenure e.g. for tenure < 12 months: assign a tenure group if 1-12

```python
copy.Tenure.max()
```

```python
labels = [f"{i+1} - {i+12}" for i in range(0, 61, 12)]
copy['tenure_group'] = pd.cut(copy.Tenure, range(1, 80, 12), right=False, labels=labels)
copy['tenure_group'].value_counts()
```

```python
copy.drop(columns= ['CustomerID','Tenure'], axis=1, inplace=True)
copy.head()
```

```python
copy.describe()
```

# Build a corelation of all predictors with 'Churn' 

```python
plt.figure(figsize=(25,8))
copy.drop('tenure_group',axis=1).corr()['Churn'].sort_values(ascending = False).plot(kind='bar')
plt.xticks(fontsize =20)
plt.yticks(fontsize=20)
plt.show()
```

From this we can see that:

1.High Churn is seen incase of Complaints, Marital Status(Single), Distance of Warehouse from home, low tier cities and surprisingly in high satisfaction score also.

2.Low Churn is seen incase of Cashback amounts, Recency, Marital Status(married) and Tenure.

3.Factors such as HoursSpendOnApp, CouponUsed, OrderAmountHikeFromlastYear, NumberofAddress and Order counts have almost no impact on Churn.

Which is also evident from the Heatmap below.

# Correlation Matrix for the Customer Dataset

```python
plt.figure(figsize=(30,15))
sns.heatmap(copy.drop('tenure_group',axis=1).corr(), annot=True)
plt.title("Correlation Matrix for the Customer Dataset")
plt.show()
```

# Insights after analysing the data:-

1. Customers who are male and single tend to have a high churn rate.
2. Customers who have filed a complaint, are from low tier cities or have a large distance of warehouse from their home also have a high churn rate.
3. Maximum number of customers spend 3 hours on their app with 99.28% of customers spending 2-4 hours on their app.
4. Most of the people use debit card and credit card as their mode of payment with COD(Cash On Delivery) being the least used.
5. Customers with longer Tenure, high cashback amounts or low recency have comparatively lower churn rate.
6. Also people with longer tenure have more cashback amount, which is obvious.

So if we were to create a model to predict whether a new customer is going to churn or not the top 5 features we can use for the model are:

- Tenure

- Cashback Amount

- Distance from Warehouse to home

- Complaint 

- Recency

- Marital Status (Single/Married)
